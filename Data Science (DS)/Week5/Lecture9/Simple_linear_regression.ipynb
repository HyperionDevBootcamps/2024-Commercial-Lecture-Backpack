```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score



import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = load_diabetes(as_frame=True).data
```

```python
df.head()
```

```python
df = df[['s1', 's2']]
```

```python
plt.figure()
sns.scatterplot(data=df, x='s1', y='s2')
plt.xlabel('s1 (total serum cholesterol)')
plt.ylabel('s2 (low-density lipoproteins)')
plt.title('Scatter Plot of s1 vs. s2')
plt.show()
plt.close()
```

```python
# Preparing the data for linear regression
X = df['s1'].values.reshape(-1, 1)
y = df['s2'].values
```

```python
# Creating and fitting the linear regression model
simple_model = LinearRegression()
simple_model.fit(X, y)
```

```python
# Making predictions
y_pred = simple_model.predict(X)
```

```python
plt.figure()
sns.scatterplot(data=df, x='s1', y='s2')
plt.plot(X, y_pred, color='red')
plt.xlabel('s1 (total serum cholesterol)')
plt.ylabel('s2 (low-density lipoproteins)')
plt.title('Linear Regression Line')
plt.show()
plt.close()
```

```python
# Evaluating the model
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Evaluation Metrics:")
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2) Score:", r2)
```
